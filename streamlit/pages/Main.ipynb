```python
import pandas as pd
import json
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.figure(figsize=(5,3),dpi=300)
plt.xlabel('Time (mins)')
plt.ylabel('Emotion intensity')

data = json.load(open('../../lang-reveal.json'))
scores = data['language']['predictions']

df = pd.DataFrame(columns=['Word', 'Joy', 'Sadness'])

joy_scores = []
sad_scores = []

for i in range(0, len(scores)):
    curr = scores[i]['emotions']
    for j in range(0, len(curr)):
        if curr[j]['name'] == 'Sadness':
            sad_scores.append(curr[j]['score'])
        elif curr[j]['name'] == 'Joy':
            joy_scores.append(curr[j]['score'])

x = np.arange(0, 60, 60/len(joy_scores))

for i in range(0, len(sad_scores)):
    df_temp = pd.DataFrame()
    df.append()

plt.plot(x, joy_scores, marker='.', markersize=4, label='Joy')
plt.plot(x, sad_scores, marker='.', markersize=4, label='Sadness')
plt.legend()
plt.show()
```
