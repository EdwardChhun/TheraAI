```python
import pandas as pd
import json
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = json.load(open('../../lang-reveal-sad.json'))
scores = data['language']['predictions']

emotions = {'Sad':'ğŸ˜¢','Neutral': 'ğŸ˜','Happy':'ğŸ˜€'}
df = pd.DataFrame(columns=['Word', 'Score', 'Emotion'])

for i in range(0, len(scores)):
    entry = []
    entry.append(scores[i]['text'])

    joy = 0
    sadness = 0
    for j in range(0, len(scores[i]['emotions'])):
        if scores[i]['emotions'][j]['name'] == 'Joy':
            joy = scores[i]['emotions'][j]['score']
        if scores[i]['emotions'][j]['name'] == 'Sadness':
            sadness = scores[i]['emotions'][j]['score']
        
    diff = joy - sadness
    entry.append(diff)

    if diff > 0.0:
        entry.append('ğŸ˜€')
    elif diff == 0.0:
        entry.append('ğŸ˜')
    else:
        entry.append('ğŸ˜¢')

    df_temp = pd.DataFrame([entry])

    df = pd.concat([df, df_temp], ignore_index=True)

df.to_csv('emotion_over_time_sad.csv', index=False)

```
